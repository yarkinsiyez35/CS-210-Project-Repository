IMPORTING LIBRARIES

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd 
```

SENDING REQUEST

```python
url = "https://hsgm.saglik.gov.tr/tr/haberler/yetkilendirilmis-covi-d-19-tani-laboratuvarlari.html"

content = requests.get(url)
content = content.text
soup = BeautifulSoup(content, 'html.parser')
```

GETTING THE TABLE

Finding the column names 

```python
header = soup.find("div", {"class": "articleBody"}).find("tbody").find("tr").find_all("td")
print(header)
```

```python
headerText = []
for i in header:
    headerText.append(i.text)
    
print(headerText)
```

Finding the content of the table 

```python
table = []
for entry in soup.find("div", {"class": "articleBody"}).find("tbody").find_all("tr"):
    contentOfTable = entry.findAll("td")
    contentOfTableText =[]
    for i in contentOfTable:
        contentOfTableText.append(i.text)
    table.append(contentOfTableText)

table = table[ 1:]    

rows = []

for i in range(len(table)):
    row = dict(zip(headerText, table[i]))
    rows.append(row)

```

Creating the dataframe 

```python
AuthorizedLabDataframe = pd.DataFrame(rows)
AuthorizedLabDataframe.head()
```

Storing the dataframe as an excel file 

```python
AuthorizedLabDataframe.to_excel("Authorized_Lab_Data.xlsx", index = False)
```
